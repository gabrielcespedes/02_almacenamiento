## **Conexión Cliente GCS**

```python
from google.cloud import storage
import os

# 1. Definimos la ruta del archivo JSON (asegúrate de que "account.json" esté en la misma carpeta)
ruta_credenciales = "account.json"

# 2. Validación de seguridad (Excelente práctica para enseñar)
if not os.path.exists(ruta_credenciales):
    print(f"ERROR: No se encontró el archivo de credenciales en la ruta: {ruta_credenciales}")
else:
    # 3. Variable de entorno
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = ruta_credenciales

    try:
        # 4. El cliente toma credenciales del entorno automáticamente
        storage_client = storage.Client()
        print("¡Éxito! Conectado a Google Cloud Storage.")
        
    except Exception as e:
        print(f"Error al conectar con GCP: {e}")
```
