## **Conexión Cliente GCS**

```python
from google.cloud import storage
import os

# 1. Definimos la ruta del archivo JSON (asegúrate de que "account.json" esté en la misma carpeta)
ruta_credenciales = "account.json"

# 2. Validación de seguridad (Excelente práctica para enseñar)
if not os.path.exists(ruta_credenciales):
    print(f"ERROR: No se encontró el archivo de credenciales en la ruta: {ruta_credenciales}")
else:
    # 3. Variable de entorno
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = ruta_credenciales

    try:
        # 4. El cliente toma credenciales del entorno automáticamente
        storage_client = storage.Client()
        print("¡Éxito! Conectado a Google Cloud Storage.")
        
    except Exception as e:
        print(f"Error al conectar con GCP: {e}")
```

## **Creación de Bucket en GCP**

```python
bucket_name = "datalake-bsg-2026"

try:
    bucket = storage_client.create_bucket(bucket_name, location = "US-CENTRAL1")
    print(f"Bucket {bucket.name} creado en {bucket.location}.")
except Exception as e:
    print(e)
```

## **Subir un Archivo (Blob)**

```python
bucket = storage_client.get_bucket(bucket_name)
blob = bucket.blob("zona_bronze/sales/data_sales.csv")

# Subir desde archivo local
blob.upload_from_filename("../data/sales.csv")

print("Archivo subido a GCS.")
```

## **Subir desde String (en memoria)**

```python
texto_data = "id,nombre,valor\n1,A,100\n2,B,200"

blob = bucket.blob("zona-silver/resumen.csv")
blob.upload_from_string(texto_data, content_type="text/csv")

print("CSV en memoria subido.")
```

## **Listar y Filtrar Blobs**

```python
blobs = storage_client.list_blobs(bucket_name, prefix = "zona_bronze/")

for blob in blobs:
    print(f"Blob: {blob.name}, Size: {blob.size}")
```

## **Descargar Archivos**

```python
blob = bucket.blob("zona_bronze/sales/data_sales.csv")
blob.download_to_filename("../data/descarga_gcs_sales.csv")

print("Descarga completada.")
```
